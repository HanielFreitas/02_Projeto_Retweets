```
import twitter
from pyspark.sql.functions import *
import pandas as pd

auth = twitter.oauth.OAuth(OAUTH_TOKEN,OAUTH_TOKEN_SECRET,CONSUMER_KEY,CONSUMER_SECRET)

twitter_api = twitter.Twitter(auth=auth)
```

```
q = 'Brasil'

count = 100

search_result = twitter_api.search.tweets(q=q,count=count)

statuses = search_result['statuses']

for _ in range(5):
    print('Tamanho do status',len(statuses))
    try:
        next_result = search_result['search_metadata']['next_results']
    except KeyError as e:
        break
        
    kwargs = dict([kv.split('=')for kv in unquote(next_result[1:]).split("&")])
    
    search_result = twitter_api.search.tweets(**kwargs)
    
    statuses += search_result['statuses']

status_texts = [status['text']
               for status in statuses]

screen_names = [user_mention['screen_name']
               for status in statuses
                   for user_mention in status['entities']['user_mentions']]

hashtags = [hashtag['text']
           for status in statuses
               for hashtag in status['entities']['hashtags']]

words = [w
         for t in status_texts
             for w in t.split()]

retweets = [
    
    #Armazene uma tupla com esses 3 valores
    (status['retweet_count'],
    status['retweeted_status']['user']['screen_name'],
    status['text'])
    
    #.. for each status
    for status in statuses
        if 'retweeted_status' in status.keys()
]
df = spark.createDataFrame(retweets)
pdf = pd.DataFrame(retweets, columns=['Count','Usuario do Twitter','Text'])
df = spark.createDataFrame(pdf)
display(df.orderBy("Count", ascending = 0).limit(30))
```
